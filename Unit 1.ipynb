```python
import plotly.graph_objs as go
```

# Line graph

```python
trace = go.Line(x=[1,2,3,4], y=[11,12,13,21])
```

```python
trace
```

```python
trace
go.Figure(trace)
```

```python
go.Figure(data=trace, layout=go.Layout(title="Line Graph"))
```

```python
x = [12,23,16,34]
y = [32,24,15,23]
trace = go.Scatter(x=x, y=y, mode="markers")
go.Figure(trace)
```

```python
x = [12,23,16,34]
y = [32,24,15,23]
trace = go.Scatter(x=x, y=y, mode="markers+lines")
go.Figure(trace)
```

```python
x = [12,23,16,34]
y = [32,24,15,23]
trace = go.Scatter(x=x, y=y, mode="lines")
go.Figure(trace)
```

```python
x = [12,23,16,34]
y = [32,24,15,23]
trace1 = go.Scatter(x=x, y=y, mode="lines", name="one")
trace2 = go.Scatter(x=[1,2,3,4], y=[3,5,6,7], mode="markers", name="two")
trace3 = go.Scatter(x=[21,13,16,18], y=[12,43,12,23], mode="text+markers", name="three", text="three")
go.Figure(data = [trace1,trace2,trace3])
```

```python
import numpy as np
```

```python
x = np.random.randint(10,90,50)
y = np.random.randint(20,220,50)
trace = go.Scatter(x=x, y=y, mode="markers", name="two")
go.Figure(trace)
```

```python
trace = go.Scatter(x=x[:20],y=y[:20], mode="markers")
go.Figure(trace)
```

```python
trace = go.Scatter(x=x[:20],y=y[:20], mode="markers", marker_size = np.random.randint(20,50,20))
go.Figure(trace, layout=go.Layout(title="Bubble Graph"))
```

```python
trace = go.Scatter(x=[23,24,32,26], y=[43,54,35,22], marker=dict(size=[10,30,60,90]), mode='markers')
go.Figure(trace)
```
